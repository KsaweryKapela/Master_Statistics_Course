```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
sleep_hours = [5, 5.5, 6, 6, 7, 7, 7.5, 8, 8.5, 9]
dollars = [47, 53, 52, 44, 39, 49, 50, 38, 43, 40]

plt.plot(sleep_hours, dollars,'ko',  markerfacecolor='k')
plt.xlabel('Hours of sleep')
plt.ylabel('Fijian dollars')
plt.show()
```

```python
## least-squares fitting

design_matrix = np.vstack((np.ones(10), sleep_hours)).T
print(design_matrix)

beta = np.linalg.lstsq(design_matrix, dollars, rcond=None)[0]
print(beta)

predicted_values = design_matrix@beta
```

```python

plt.plot(sleep_hours, dollars, 'ko', markerfacecolor='k')
plt.plot(sleep_hours, predicted_values, 's-')

for i in range(10):
    plt.plot([sleep_hours[i], sleep_hours[i]], [dollars[i], predicted_values[i]], 'm--')
    

plt.legend(('Data (y)','Model ($\^{y}$)','Residuals'))

plt.xlabel('Hours of sleep')
plt.ylabel('Fijian dollars')
plt.show()
```

```python
slope,intercept, r, p, std_err = stats.linregress(sleep_hours,dollars)
print(intercept,slope)
print(beta)
print(p)
```
