```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

h_0 = 0
n = 200
population_mean = -0.5
std_size = 30

std_value = np.linspace(0.1, 3, std_size)
t_value = np.zeros(std_size)
p_value = np.zeros(std_size)

for i in range(std_size):
    data  = np.random.normal(population_mean, std_value[i], n) 
    t_value[i], p_value[i] = stats.ttest_1samp(data, h_0)
```

```python
fig, ax = plt.subplots(1,3, figsize=(20, 8))
ax[0].plot(std_value, t_value, 'o')
ax[0].set_xlabel('Standard Deviation')
ax[0].set_ylabel('T')
 
ax[1].plot(std_value, p_value, 'o')
ax[1].set_xlabel('Standard Deviation')
ax[1].set_ylabel('P')
 
ax[2].plot(t_value, p_value, 'o')
ax[2].set_xlabel('T')
ax[2].set_ylabel('P')
 
plt.tight_layout()
plt.show()
```
